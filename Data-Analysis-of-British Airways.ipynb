```python
fig, ax = plt.subplots(figsize=(7, 6))
counts = data1['Recommended'].value_counts().values
labels = data1['Recommended'].value_counts().index


ax.bar(labels, counts,color=['#006400','#461B7E'],label = ("no","yes"))


for i, v in enumerate(counts):
    ax.text(x=labels[i], y=v + 0.01 * counts.max(), s="{:.0%}".format(v / counts.sum()), color='black', size=12, ha='center')

plt.xlabel('Recommended', fontsize=16)
plt.xticks(labels, size=19)
plt.legend(title = 'Recommended',shadow=True,fontsize="15")
plt.show()
```

## In above Graph we can see that **62%** Reviewers are give **No** Recommended and other hand **38%** Reviewers are give **Yes** recommended

```python
def add_percentage_labels(ax):
    total = len(data1)

    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')

sns.set_theme(style='whitegrid', palette='RdBu_r')
plt.figure(figsize=(15, 7))

ax = sns.countplot(data=data1, x='Recommended', hue='Value For Money', saturation=1.100, edgecolor='black')

add_percentage_labels(ax)

plt.title('Recommended and Value For Money', fontsize=17)
plt.xlabel('Recommended', fontsize=17)
plt.legend(title="Value For Money", fontsize="12")
plt.xticks(size=17)
plt.show()
```

In the above graph The Total **62.4%** of the Reviewers are **Recommend no** And **37.7%** Reviewers are **recommended yes**,most of the reviewers are not happy with the services of the airlines.
