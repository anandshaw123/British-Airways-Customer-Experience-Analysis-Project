## **Drop Unnecessary Columns**


```python
columns = ['Date Flown','Aircraft Type','Inflight Entertainment','Wifi & Connectivity','Ground Service','Route','Food & Beverages']

df.drop(columns,axis = 1,inplace=True)
```

```python
# Drop Null Values
df.dropna(subset=['Type_of_Travellers'],axis=0,inplace=True)
df.dropna(subset=['Cabin Staff Service'],axis=0,inplace=True)
df.dropna(subset=['Seat_Types'],axis=0,inplace=True)
```
